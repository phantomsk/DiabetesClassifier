```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Diabetes_012'], bins=3, discrete=True)
plt.title('Distribution of Diabetes Status')
plt.xlabel('Diabetes Status (0 = No, 1 = Prediabetes, 2 = Diabetes)')
plt.ylabel('Count')
plt.xticks([0, 1, 2])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['BMI'], bins=50)
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(x=data['BMI'])
plt.title('Box Plot of BMI (with Outliers)')
plt.xlabel('BMI')
plt.tight_layout()
plt.show()
```

```python
percentile25 = data['BMI'].quantile(0.25)
percentile75 = data['BMI'].quantile(0.75)
IQR = percentile75 - percentile25
lower_bound = percentile25 - 1.5 * IQR
upper_bound = percentile75 + 1.5 * IQR
```

```python
new_health_indicators = data[data['BMI'] < upper_bound].copy()

new_health_indicators["BMI"].describe()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x=data['BMI'])
plt.title("BMI with Outliers")
plt.xlabel("BMI")

plt.subplot(1, 2, 2)
sns.boxplot(x=new_health_indicators['BMI'])
plt.title("BMI after Removing High Outliers")
plt.xlabel("BMI")

plt.tight_layout()
plt.show()
```

```python
new_health_indicators["BMI"].groupby(new_health_indicators["Diabetes_012"]).median()
```

```python
summary_cols = ['Diabetes_012', 'BMI', 'HighBP', 'HighChol', 'CholCheck', 
                'Smoker', 'Stroke', 'HeartDiseaseorAttack', 'Age']

summary_stats = new_health_indicators[summary_cols].describe().T
summary_stats
```

```python
plt.figure(figsize=(10, 8))
corr = new_health_indicators.corr(numeric_only=True)

sns.heatmap(corr, cmap='coolwarm', center=0, square=False)
plt.title('Correlation Heatmap of Health Indicators')
plt.tight_layout()
plt.show()
```

```python
mean_bmi = new_health_indicators.groupby('Diabetes_012')['BMI'].mean()

plt.figure(figsize=(8, 6))
sns.barplot(x=mean_bmi.index, y=mean_bmi.values)
plt.title('Average BMI by Diabetes Status')
plt.xlabel('Diabetes Status (0 = No, 1 = Prediabetes, 2 = Diabetes)')
plt.ylabel('Mean BMI')
plt.tight_layout()
plt.show()
```

```python
# Proportion with diabetes (status==2) in each age group
age_diab = new_health_indicators.groupby('Age')['Diabetes_012'].apply(
    lambda x: np.mean(x == 2)
).reset_index(name='DiabetesPrevalence')

plt.figure(figsize=(8, 6))
sns.lineplot(data=age_diab, x='Age', y='DiabetesPrevalence', marker='o')
plt.title('Prevalence of Diabetes (Status = 2) by Age Group')
plt.xlabel('Age Category')
plt.ylabel('Proportion with Diabetes')
plt.tight_layout()
plt.show()
```
