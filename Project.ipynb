```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
diabetes = pd.read_csv('diabetes.csv')
```

```python
corrs = diabetes.corr()
corrs
```

```python
diabetes.describe()
```

```python
fig, ax = plt.subplots(1)
#temp = diabetes
#temp['Diabetes_012'] = temp.Diabetes_012.replace({0:'Non-Diabetic', 1:'Pre-Diabetic', 2:'Diabetic'})
sns.histplot(data=diabetes, bins = 3, x='Diabetes_012', stat = 'percent')
ax.set_xlabel('Stage of Diabetes')
ax.set_ylabel('Percentage of Population Sampled')
ax.set_title('Percentage of Sampled Population by Diabetes Stage')
```

```python
diabetics = diabetes[diabetes['Diabetes_012'] == 2]
non_diabetics = diabetes[(diabetes['Diabetes_012'] == 0) | (diabetes['Diabetes_012'] == 1)]
fig, ax = plt.subplots(1)
sns.histplot(bins = 5, data = diabetics, x='GenHlth', stat='percent')
ax.set_xlabel('General Health Score')
ax.set_ylabel('Percent of Diabetics')
ax.set_title('Percentage of Diabetics per General Health Score')
```

```python
fig, ax = plt.subplots(1)
sns.histplot(bins = 5, data = non_diabetics, x='GenHlth', stat='percent')
ax.set_xlabel('General Health Score')
ax.set_ylabel('Percent of Non-Diabetics')
ax.set_title('Percentage of Non-Diabetics per General Health Score')
```

```python
fig, ax = plt.subplots(1)
temp = diabetes
temp['Diabetes_012'] = diabetes.Diabetes_012.replace({0:'Non-Diabetic', 1:'Pre-Diabetic', 2:'Diabetic'})
sns.boxplot(data=temp, x = 'Diabetes_012', y='BMI')
ax.set_title('BMI by Diabetic Stage')
ax.set_xlabel('Diabetic Stage')
```
